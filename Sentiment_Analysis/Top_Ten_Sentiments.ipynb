### Defines the top ten functions, which will gather the top 10 best and worst reviews according to different sentiment scores 

```python
#defines a function that outputs the top 10 most positive and negative sentiment reviews by compound and neg/pos scores
def top_ten_comp(df):
    np.set_printoptions(threshold=sys.maxsize)
    
    #Compound sentiment - positive
    print('Top 10 Most Positive Reviews by Compound Sentiment')
    print()
    df = df.sort_values(by = 'comp_sent', ascending = True).reset_index(drop = True) # inplace = True, 
    for i in range(10):
        #df.sort_values(by = 'comp_sent', ascending = True).reset_index() # inplace = True, 
        print(str(i+1)+ '.) ' + df['full_text'][i])
        print()
    
    #Compound Sentiment - negative
    print()
    print('Top 10 Most Negative Reviews by Compound Sentiment')
    print()
    df = df.sort_values(by = 'comp_sent', ascending = False).reset_index(drop = True) # inplace = True,
    for i in range(10):
        print(str(i+1)+ '.) ' + df['full_text'][i])
        print()
```

```python
def top_ten_pn(df):
    np.set_printoptions(threshold=sys.maxsize)
    
    #Pos and Neg Sentiment
    print('Top 10 Most Positive and Negative Reviews by Positive and Negative Sentiment')
    print()
    
    #positive sentiment
    print('Top 10 Most Positive Reviews by Compound Sentiment')
    print()
    df = df.sort_values(by = 'pos_sent', ascending = True).reset_index(drop = True) # inplace = True, 
    for i in range(10):
        #df.sort_values(by = 'comp_sent', ascending = True).reset_index() # inplace = True, 
        print(str(i+1)+ '.) ' + df['full_text'][i])
        print()
    
    #Compound Sentiment - negative
    print()
    print('Top 10 Most Negative Reviews by Compound Sentiment')
    print()
    df = df.sort_values(by = 'neg_sent', ascending = False).reset_index(drop = True) # inplace = True,
    for i in range(10):
        print(str(i+1)+ '.) ' + df['full_text'][i])
        print()
```
